<h3>Моделирование:

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import  MinMaxScaler 
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
import numpy as np
```

Дерево решений (DecisionTreeClassifier) – это инструмент прогнозного моделирования, применяемого в ряде различных областей. Как правило, они строятся с помощью алгоритмического подхода, который определяет способы разделения набора данных на основе различных условий. Основная идея алгоритма заключается в построении бинарного дерева, в котором каждый внутренний узел представляет собой условие на признаках, а листья — конечный результат работы алгоритма (например, принадлежность к определенному классу).

Алгоритм случайного леса (Random Forest) — универсальный алгоритм машинного обучения, суть которого состоит в использовании ансамбля решающих деревьев. Само по себе решающее дерево предоставляет крайне невысокое качество классификации, но из-за большого их количества результат значительно улучшается. Также это один из немногих алгоритмов, который можно использовать в абсолютном большинстве задач.

GridSearchCV – это очень мощный инструмент для автоматического подбирания параметров для моделей машинного обучения. GridSearchCV находит наилучшие параметры, путем обычного перебора: он создает модель для каждой возможной комбинации параметров. 

MinMaxScaler – библиотека Scikit-learn, позволяющая произвести Нормализацию (Normalization) данных перед использованием в Модели (Model) Машинного обучения (ML), то есть приведение Числовых переменных (Numeric Variable) к диапазону от 0 до 1.

SVC - классификатор опорных векторов, представляет собой контролируемый алгоритм машинного обучения, обычно используемый для задач классификации

KNeighborsClassifier - простейший метрический классификатор, основанный на оценивании сходства объектов. Классифицируемый объект относится к тому классу, которому принадлежат ближайшие к нему объекты обучающей выборки.

- Кодируем категориальные признаки
- Разбиваем данные на тренировочную тестовую и валидационную выборки

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
df.gender = le.fit_transform(df.gender)
df.oral = le.fit_transform(df.oral)
df.tartar = le.fit_transform(df.tartar)
```

```python
y = df.smoking
df.drop(['smoking'], axis = 1, inplace= True)
```

```python
X = df
```

```python

from sklearn.model_selection import train_test_split


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)

X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.25, random_state=1) 
```

```python
# scaler = MinMaxScaler()
# scaler.fit(X_train)
# X_train = scaler.transform(X_train)
# X_test = scaler.transform(X_test)
# X_val = scaler.transform(X_val)
```

В качестве первой модели используем дерево решений, используем GridSearchCV для подбора гиперпараметров

```python
grid = {
        'max_depth': (range(1, 10, 1)),
        'max_features': (range(1, 5, 1))}
gridsearch = GridSearchCV(DecisionTreeClassifier(random_state=42), grid, scoring='roc_auc', n_jobs=-1)
```

```python
gridsearch.fit(X_train, y_train)
gridsearch.best_params_
```

```python
clf = DecisionTreeClassifier(random_state=42, max_depth = 9, max_features = 4 )
clf.fit(X_train, y_train)
```

```python
predicted = clf.predict(X_train)
print(accuracy_score(y_train, predicted))
```

```python
predicted = clf.predict(X_test)
print(accuracy_score(y_test, predicted))
```

```python

predicted = clf.predict(X_val)
print(accuracy_score(y_val, predicted))
```

Вторая модель случайный лес

```python
rf_model = RandomForestClassifier(n_estimators = 1000)
rf_model.fit(X_train, y_train)
```

```python
predicted = rf_model.predict(X_train)
print(accuracy_score(y_train, predicted))
```

```python

predicted = rf_model.predict(X_test)
print(accuracy_score(y_test, predicted))
```

```python
predicted = rf_model.predict(X_val)
print(accuracy_score(y_val, predicted))
```

```python
print(classification_report(y_val, predicted))
```

```python
importances = (RandomForestClassifier().fit(X_train, y_train).feature_importances_)
indices = np.argsort(importances)[::-1]
feature_names = df.columns
x =importances[indices]
y = feature_names[indices]
plt.figure(figsize=(20,6))
sns.set_theme(style="ticks", font_scale=1.5)
sns.barplot(x=x[0:15], y=y[0:15]);
```

Ансамбь моделей

```python
from sklearn.ensemble import VotingClassifier

dtc = DecisionTreeClassifier()
svc = SVC (probability=True)
knn = KNeighborsClassifier()
```

```python
voting_clf = VotingClassifier(
    estimators=[('svc', svc), ('dtc', dtc), ('knn', knn)],
    voting='hard')
```

VotingClassifier — это оценщик с машинным обучением, который обучает различные базовые модели или оценщики и делает прогнозы на основе агрегирования результатов каждого базового оценщика . Критерии агрегирования могут быть объединены решением голосования для каждого вывода оценщика.

```python
voting_clf.fit(X_train, y_train)
```

```python
predicted=voting_clf.predict(X_train)
print(accuracy_score(y_train, predicted))
```

```python
predicted=voting_clf.predict(X_test)
print(accuracy_score(y_test, predicted))
```

```python
predicted=voting_clf.predict(X_val)
print(accuracy_score(y_val, predicted))
```
